```python
from qaoamis import QAOAMIS

mis = QAOAMIS()
penalty = mis.n

mis.addNodes(4)

mis.addEdges([
    (0, 1), (1, 2), (2, 0), (3, 0)
])

mis.visualize()
```

```python
mis.buildQUBO(penalty=penalty)
mis.mis_brute_force(show_graph=True, print_expected_energy_bar=True)
```

```python
mis.buildCostHamiltonian(penalty=penalty)
circuit = mis.buildQAOAAnsatz(layers=3)

print(circuit)
```

```python
beta, gamma = mis.find_optimal_parameters()
```

```python
print(beta, gamma)
mis.qaoa_evaluate(beta, gamma, shots=2 ** 16, print_count=True)
```
