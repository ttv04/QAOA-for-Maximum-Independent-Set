```python
from qaoamis import QAOAMIS

mis = QAOAMIS()

mis.addNodes(6)

mis.addEdges(
    [(0, 1), (2, 3), (4, 5)]
)

mis.visualize()
```

```python
mis.buildQUBO(penalty=mis.n)
mis.mis_brute_force(show_graph=True, print_expected_energy_bar=True)
```

```python
mis.buildCostHamiltonian(penalty=mis.n)
mis.buildQAOAAnsatz(layers = 3, draw_circuit=True)
beta, gamma = mis.find_optimal_parameters()
```

```python
mis.qaoa_evaluate(beta, gamma, shots=2 ** 16, print_count=True)
```
